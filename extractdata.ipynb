```python
import os
from ecl.summary import EclSum
import numpy as np
from datetime import date, datetime, timedelta
import sys
import matplotlib.pyplot as plt

filename = './three wells/XFLOW2'
strModelName=filename
tSummaryData=EclSum(strModelName+'.UNSMRY')
#print(tSummaryData.keys())
afTime=tSummaryData.numpy_vector("TIME")
year = '2023'
strt_date = date(int(year),1,1)
afTimeD = [strt_date + timedelta(days=diff -1) for diff in afTime]
afTimeNPX = np.array(afTimeD, dtype='datetime64[s]')
afOPR_2wells=tSummaryData.numpy_vector("FOPR")
afOPT_2wells=tSummaryData.numpy_vector("FOPT")
afWIR_2wells=tSummaryData.numpy_vector("FWIR")
afWIT_2wells=tSummaryData.numpy_vector("FWIT")
crossflow_top=tSummaryData.numpy_vector("CWFR:I:1,1,1")
crossflow_bottom=tSummaryData.numpy_vector("CWFR:I:1,1,3")
afCWFRI_top=tSummaryData.numpy_vector("CWFR:INJ:1,1,1")
afCWFRI_bottom=tSummaryData.numpy_vector("CWFR:INJX:1,1,3")
afWBHP_Prod_2wells=tSummaryData.numpy_vector("WBHP:PROD")
afWBHP_Inj_2wells=tSummaryData.numpy_vector("WBHP:INJ")
afBPR_top_2wells=tSummaryData.numpy_vector("BPR:1,1,1")
afBPR_bottom_2wells=tSummaryData.numpy_vector("BPR:1,1,3")

filename = './one well/XFLOW'
strModelName=filename
tSummaryData=EclSum(strModelName+'.UNSMRY')
#print(tSummaryData.keys())
afOPR_Xflow=tSummaryData.numpy_vector("FOPR")
afOPT_Xflow=tSummaryData.numpy_vector("FOPT")
afWIR_Xflow=tSummaryData.numpy_vector("FWIR")
afWIT_Xflow=tSummaryData.numpy_vector("FWIT")
afCWFR_Xflow=tSummaryData.numpy_vector("CWFR:INJ:1,1,1")
afCWFRX_Xflow=tSummaryData.numpy_vector("CWFR:INJ:1,1,3")
afWBHP_Prod_Xflow=tSummaryData.numpy_vector("WBHP:PROD")
afWBHP_Inj_Xflow=tSummaryData.numpy_vector("WBHP:INJ")
afBPR_top_Xflow=tSummaryData.numpy_vector("BPR:1,1,1")
afBPR_bottom_Xflow=tSummaryData.numpy_vector("BPR:1,1,3")

filename = './one well/NOXFLOW'
strModelName=filename
tSummaryData=EclSum(strModelName+'.UNSMRY')
#print(tSummaryData.keys())
afTime=tSummaryData.numpy_vector("TIME")
year = '2023'
strt_date = date(int(year),1,1)
afTimeD = [strt_date + timedelta(days=diff -1) for diff in afTime]
afTimeNP = np.array(afTimeD, dtype='datetime64[s]')
afOPR_NXflow=tSummaryData.numpy_vector("FOPR")
afOPT_NXflow=tSummaryData.numpy_vector("FOPT")
afWIR_NXflow=tSummaryData.numpy_vector("FWIR")
afWIT_NXflow=tSummaryData.numpy_vector("FWIT")
afCWFR_NXflow=tSummaryData.numpy_vector("CWFR:INJ:1,1,1")
afWBHP_Prod_NXflow=tSummaryData.numpy_vector("WBHP:PROD")
afWBHP_Inj_NXflow=tSummaryData.numpy_vector("WBHP:INJ")
afBPR_top_NXflow=tSummaryData.numpy_vector("BPR:1,1,1")
afBPR_bottom_NXflow=tSummaryData.numpy_vector("BPR:1,1,3")


#FOPR
plt.figure(figsize=(12, 6))
plt.plot(afTimeNPX,afOPR_2wells,label='Xflow with controlled valve',color='red')
plt.plot(afTimeNPX,afOPR_Xflow,label='Xflow without controlled valve',color='blue')
plt.plot(afTimeNP,afOPR_NXflow,label='NoXflow',color='green')
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b-%y'))
plt.xlabel('Time')
plt.ylabel('OPR[SM3/DAY]')
plt.title(f'Oil Production Rate[SM3/DAY]')
plt.legend()
plt.show()

#FOPT
plt.figure(figsize=(12, 6))
plt.plot(afTimeNPX,afOPT_2wells,label='Xflow with controlled valve',color='red')
plt.plot(afTimeNPX,afOPT_Xflow,label='Xflow without controlled valve',color='blue')
plt.plot(afTimeNP,afOPT_NXflow,label='NoXflow',color='green')
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b-%y'))
plt.xlabel('Time')
plt.ylabel('OPT[SM3]')
plt.title(f'Oil Production Total[SM3]')
plt.legend()
plt.show()

#FWIR
plt.figure(figsize=(12, 6))
plt.plot(afTimeNPX,afWIR_2wells,label='Xflow with controlled valve',color='red')
plt.plot(afTimeNPX,afWIR_Xflow,label='Xflow without controlled valve',color='blue')
plt.plot(afTimeNP,afWIR_NXflow,label='NoXflow',color='green')
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b-%y'))
plt.xlabel('Time')
plt.ylabel('WIR[SM3/DAY]')
plt.title(f'Water Injection Rate[SM3/DAY]')
plt.legend()
plt.show()

#FWIT
plt.figure(figsize=(12, 6))
plt.plot(afTimeNPX,afWIT_2wells,label='Xflow with controlled valve',color='red')
plt.plot(afTimeNPX,afWIT_Xflow,label='Xflow without controlled valve',color='blue')
plt.plot(afTimeNP,afWIT_NXflow,label='NoXflow',color='green')
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b-%y'))
plt.xlabel('Time')
plt.ylabel('WIT[SM3]')
plt.title(f'Water Injection Total[SM3]')
plt.legend()
plt.show()

#Block Pressure
plt.figure(figsize=(12, 6))
plt.plot(afTimeNPX,afBPR_top_2wells,label='1,1,1 with controlled valve',color='red')
plt.plot(afTimeNPX,afBPR_bottom_2wells,label='1,1,3 with controlled valve',linestyle='dashed',color='red')
plt.plot(afTimeNPX,afBPR_top_Xflow,label='1,1,1 without controlled valve',color='blue')
plt.plot(afTimeNPX,afBPR_bottom_Xflow,label='1,1,3 without controlled valve',linestyle='dashed',color='blue')
plt.plot(afTimeNP,afBPR_top_NXflow,label='1,1,1 NoXflow',color='green')
plt.plot(afTimeNP,afBPR_bottom_NXflow,label='1,1,3 NoXflow',linestyle='dashed',color='green')
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b-%y'))
plt.xlabel('Time')
plt.ylabel('Block Pressure [Bar]')
plt.title(f'Block Pressure [Bar]')
plt.legend()
plt.show()

#injection well BHP
plt.figure(figsize=(12, 6))
plt.plot(afTimeNPX,afWBHP_Inj_2wells,label='Xflow with controlled valve',color='red')
plt.plot(afTimeNPX,afWBHP_Inj_Xflow,label='Xflow without controlled valve',color='blue')
plt.plot(afTimeNP,afWBHP_Inj_NXflow,label='NoXflow',color='green')
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b-%y'))
plt.xlabel('Time')
plt.ylabel('BHP [Bar]')
plt.title(f'Injection well BHP [Bar]')
plt.legend()
plt.show()

#crossflow
plt.figure(figsize=(12, 6))
#plt.plot(afTimeNPX,crossflow_top,label='1,1,1 with controlled valve',color='red')
plt.plot(afTimeNPX,crossflow_bottom,label='crossflow with controlled valve', color='red')
#plt.plot(afTimeNPX,afCWFR_Xflow,label='1,1,1 without controlled valve',linestyle='dashed',color='orange')
plt.plot(afTimeNPX,afCWFRX_Xflow,label='crossflow without controlled valve', color='blue')
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b-%y'))
plt.xlabel('Time')
plt.ylabel('Water flow rate [SM3/DAY]')
plt.title(f'Cross Flow Rate')
plt.legend()
plt.show()
```
